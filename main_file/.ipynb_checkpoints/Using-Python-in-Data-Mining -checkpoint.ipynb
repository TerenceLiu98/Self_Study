## Classification: Basic Concepts, Decision Trees, and Model Evaluation 

Classification which is the task of assigning objects to one of several predefined categories, is a pervasive problem that encompasses many diverse applications. Classification techniques are most suited for predicting or describing data sets with binary or nominal categories. For instance, Example in *Introduction of  Data Mining*  4.2 shows how classification model works. 

The evaluation of the performance of a classification model is based on the counts of test records correctly and incorrectly predicted by the model. These counts are tabulated in a table known as a **confusion matrix**.

$$\begin{pmatrix} f_{11} & f_{10} \\ f_{01} & f_{00} \end{pmatrix}$$

the total number of **correct predictions** made by the model is $(f_{11} + f_{00})$ and the total number of **incorrect predictions** is $(f_{10} + f_{01})$

*where $0$ represents wrong and $1$ represents right*



### Decision tree 

As the name said, 'Decision Tree' is a tree that can make decision, for instance, whether an animal is a mammal animal or not is depends on multiple conditions. Using the Decision Tree can help us make the classification / decision rapidly. *Decision Tree also can do the regression either.*

The basic idea behind any decision tree algorithm is as follows: 
1. Select the best attribute using Attribute Selection Measures (ASM) to split the records 
2. Make that attribute a decision node and breaks the dataset into smaller subsets.
3. Starts tree building by repeating this process recursively for each child until one of the condition will match:\
    * All the tuples belong to the same attribute value.
    * There are no more remaining attributes.
    * There are no more instances. 
    
We use **Entropy** to evaluate which "target" should be the node. The concept of Entropy is generated by physics and it is a value to measure **the uncertainty of a system** . In information theory, it refers to the impurity in a group of examples and **Information Gain** computes the difference between before and after: $\Delta \text{Entropy}$.

$$\text{Entropy(i)} = -\sum_{i = 1}^m (p_i|t) log_2(p_i|t)$$

where, $p_i$ denotes the fraction of records belonging to class $i$ at a fiven node $t$; 


$$\text{Gini(t)} = 1 - \sum_{i = 0}^m [p(i|t)]^2$$
$$ \text{Classification error(t)} = 1 - \max\limits_{i}[p(i|t)]$$


```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
import os

col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
# load dataset
# file_name = 'data_set\diabetes.csv'
#os.chdir(os.path.dirname(file_name))
#trainData = pd.read_csv(os.path.basename(file_name))
pima = pd.read_csv(r"F:\Using-Python-Series\data_set\diabetes.csv", header=None, names=col_names)
pima.head()
```

Here, you need to divide given columns into two types of variables dependent (or target variables) and independent variables (or feature variables). 

```python
#split dataset in features and target variable
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = pima[feature_cols] # Features
y = pima.label # Target variable
```

To understand model performance, dividing the dataset into a training set and a test set is a good strategy.
Let's split the dataset by using function train_test_split(). You need to pass 3 parameters features, target, and test_set size.

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

#### Building Decision Tree Model 
Let's create a Decision Tree Model using Scikit-Learn

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_train)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
